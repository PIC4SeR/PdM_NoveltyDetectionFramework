```python
strings=['axBearDx_','axBearSx_','ayBearDx_','ayBearSx_','axShaft_','ayShaft_']
Data=pd.read_csv('Dataset_Low_Lubrication_5.txt',delimiter=',')
Data_pre=pd.DataFrame()
Data_pre=pd.concat([Data_pre,Data.loc[0:4]])
Data_pre=pd.concat([Data_pre,Data.loc[0:4]])
Data_pre=pd.concat([Data_pre,Data.loc[0:4]])
Data_pre.reset_index()
names=[]
for string in strings:
    for i in range(1,21):
        names.append(str(string+str(i)))
column_list=Data_pre.columns.values
dato=[]
for string in strings:
    count=0
    for element in column_list:
        if string in element:
            count=count+1
    dato.append(Data_pre.columns.get_loc(string + '1'))
    dato.append(Data_pre.columns.get_loc(string + str(count)))

New_Dataframe=pd.DataFrame()
for j in range(0,len(Data_pre),1):
    sum_abs=[]
    for i in range(0,len(dato),2):
        Data_Filtered=Data_pre.iloc[j,dato[i]:dato[i+1]] #Data isolation between indexes chosen before
        array_windowed = np.array_split(np.array(Data_Filtered.values),20) #Windowing 
        for element in array_windowed:
            sum_abs.append(np.sum(np.absolute(element))) #Absolute sum for each window
    All_Data=pd.DataFrame(np.array(sum_abs).reshape(1,-1),columns=names)
    New_Dataframe=pd.concat([New_Dataframe,All_Data],axis=0)
New_Dataframe=New_Dataframe.reset_index(drop=True)
#New_Dataframe['Timestamp']=Data_pre['Timestamp']
New_Dataframe
```

```python
label='axBearDx_7'
ntrain=5
try:
    del New_Dataframe['Timestamp']
except:
    print('Timestamp already deleted!')

#Features isolation
labels=New_Dataframe.iloc[:,New_Dataframe.columns.get_loc(label)].values
Feature_Dataframe=New_Dataframe.drop(columns=[label])
features=[]
for i in range(len(Feature_Dataframe)):
    features.append(Feature_Dataframe.iloc[i,:].values)
train_features=features[0:ntrain]
train_labels=labels[0:ntrain]
test_features=features[ntrain:]
test_labels=labels[ntrain:]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
param_grid=[
    {'n_estimators':[3,4,5,6,7],'max_features':[None,2,4,5,6,7,8],'random_state':[0]},
    {'n_estimators':[3,4,5,6,7],'max_features':[None,2,4,5,6,7,8],'random_state':[42]},
    {'bootstrap':[False],'n_estimators':[3,4,5,6,7],'max_features': [None,2,4,5,6,7,8],'random_state':[0]},
    {'bootstrap':[False],'n_estimators':[3,4,5,6,7],'max_features': [None,2,4,5,6,7,8],'random_state':[42]},
    {'bootstrap':[False],'n_estimators':[80,90,100,110,120],'max_features': [None,2,4,5,6,7,8,20,30,40,50],'random_state':[42]},
    {'n_estimators':[80,90,100,110,120],'max_features': [None,2,4,5,6,7,8,20,30,40,50],'random_state':[42]}
]
grid_search=GridSearchCV(rf,param_grid,scoring='neg_mean_absolute_error',return_train_score=True,n_jobs=-1)
grid_search.fit(train_features,train_labels)
print(grid_search.best_params_)
```

```python
cvres=grid_search.cv_results_
for mean_score,params in zip(cvres['mean_test_score'],cvres['params']):
    print(np.sqrt(-mean_score),params)
print(grid_search.best_params_)
```
